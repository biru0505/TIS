#함수#
**함수는 입력(input)에 따라 적절한 출력(output)을 반환(return)하는 것이다.**  
내제 함수 형식 :  
 함수명(입력값) => 출력  
 type(1) =>int  
 len("가나다") =>3  
 함수는 기본적으로 **과정을 저장**하여 그 과정을 반복적으로 사용할 때 사용하여 같은 코드를 작성하는 것을 **효율적으로 줄이는 것**에 목적이 있다.

##함수 만들기##
def 함수이름(임의변수):  
  수행할 문장1  
  수행할 문장2  
  ...  
  return 출력될 값  *return값이 없는 경우도 있다.*



```python
a = 1
b = a +1
#print(b)

#이것을 함수화 하기#
def mine(i):
  x = i **3
  return x
print(mine(10))
print(mine(11))
print(mine(2222))
```

---
함수를 정의 하는 키워드는 **def**이며, 함수이름(함수를 호출 할 때 사용).인자(함수 입력을 위해 사용되는 임의변수)를 의미한다.  
return키워드를 사용하여 출력을 변환한다.  
함수이름()형식을 통해 만들어낸 함수를 사용 할 수 있다.

##함수 인자의 기본값 설정##

```python
#입력한 메세지를 n번 출력하는 함수 n=?형태로 디폴트 값을 설정해 줄 수 있음
def msg(i="메세지삭성해주세요",n=2):
  for n in range(n):
    print(i)

msg("hi")
```
