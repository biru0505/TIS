#리스트 컴프리헨션과 딕셔너리 컴프리헨션#


##리스트 컴프리헨션##
for구문을 단축시켜 리스트를 생성하는 방법

1.1부터 5까지 정수 리스트를 만들어보자

```python
##직접작성하기
n_list=[]
n_list.append(1)
n_list.append(2)
n_list.append(3)
n_list.append(4)
n_list.append(5)

print(n_list)
```

```python
##for문으로 작성하기
n_list=[]
for n in range(1,6):
  n_list.append(n)

print(n_list)
```

##리스트 컴프리헨션의 형식##
>표현식 for 항목 in 순회가능한 객체

```python
n_list = [n for n in range(1,6)]
print(n_list)
```

첫 표현식은 리스트 변수의 이름이며
n은 요소의 이름을 for(반복)해주세요 in(몇회만큼) range(범위만큼)

#연습문제#
문자열로 구성된 리스트 축구,농구,배구의 각각의 원소에 "좋아해요"라는 문자열을 덧붙이시오.  
*리스트 컴프리헨션을 사용하시오.*

```python
favorite_sports = ["축구","농구","배구"]
str = [ i + '좋아해요!' for i in favorite_sports]
print(str)
```

리스트 컴프리헨션은 대괄호[] 안에 for문이 있다.
조건 표현식을 포함할 수 있다.  
  
  [표현식 for 항목 in 순회가능한 객체 if 조건]
  

#연습문제#
1과 5 사이에서 홀수 리스트를 만드는 새 컴프리헨션을 만드시오.

```python
n = [n for n in range(1,6) if n % 2 ==1]
print(n)

a = []
for n in range(1,6):
  if n %2 ==1:
    a.append(n)
print(a)
```

#딕셔너리 컴프리헨션#
형식 :  
{키 :값 for 표현식 in 순회가능 객체}

```python
dict_a = {i : 3 for i in range(1,6)}
print(dict_a)
```

if문을 사용 할 수 있다.  
{키 : 값 for 표현식 in 순회가능객체 if 조건}

```python
dict_b = {i:i**2 for i in range(10) if i**2<10}
dict_b
```
