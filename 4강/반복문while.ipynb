#while문#
---
###while문은 **특정한 조건을 만족하는 동안** 코드블록을 반복한다.  ###
---
while문은 반드시 코드블록 이전에 초기값을 설정해 주어야 한다.


```python
x= 1
while x<=5:
  print(x)
  x +=1
print(x) ##중요 (print 되는 것은 5까지지만, 실제의 x값은 6상태에 멈춰있다.)
```

---
#for 과 while의 사용처#
for에는 반복할 횟수를 알거나 특정할 경우 사용한다.
반면 while은 어떤 조건이 만족할 때까지의 횟수를 모를 때 주로 사용된다.
ex) 2를 몇번 제곱해야 1억보다 큰 값이 되는지 알아내시오.

```python
n=1
while 2**n <10**8:
  n +=1
print(n)
print(2**27)
```

#무한 루프와 break문#


```python
n =1
while False: #True면 계속 아래 코드블록을 반복할 것임으로, 무한 루프가 생성될 것
  if 2**n >= 10**8 :
    print(n)
    break
    n += 1
print(n)
```
