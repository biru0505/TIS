#배열 자료형#   
#**리스트(List)** 와 **튜플(Tuple)**#
**배열(array)**은 여러 개의 요소를 하나의 묶음으로 관리하는 자료형
리스트와 튜플 모두 **괄호**안에 **요소(element)를 콤마로 구분하여 입력**하는데, **리스트**는 **대괄호([ ])**를 사용하고,  
**튜플**은 **소괄호(( ))**를 사용합니다.

```python
# 리스트
L = [1, 2, 3, 4]
L2 = [5, 6, 7, 8]

#튜플
T = (1, 2, 3, 4)
T2 = (5, 6, 7, 8)

print(L, L2)
print(T, T2)
print(type(L))
print(type(T))
```

---
*참조 - 딕셔러니(Dictonary)는 중괄호({})를 사용한다 *

리스트와 튜플은 **어떠한 자료형도 요소로 사용 할 수 있습니다.**

```python
L = ["abc",10, ["A","B"], (1,2,3)]
print(L)
```

#**인덱싱(indexing)**과 **슬라이싱(slicing)**
리스트와 튜플 또한 문자열과 같이 인덱싱과 슬라이싱이 가능하다.  
*인덱싱형식 : 리스트명[인덱스번호], 튜플[인덱스번호]*  
*슬라이싱 형식 리스트명[시작인덱스번호 : 끝인덱스번호 : 건너뛰기갯수]*

```python
#인덱싱
L = ["abc",10, ["A","B"], (1,2,3)]
print(L[0], L[1], L[-1])

#슬라이싱
L = ["A", "B", "C", "D", "E"]
print(L[1:3])
```

#연습문제#
다음을 요소로 갖는 리스트를 만든 후,  
1,2,3,"A", "B", True, [4,5,6]  
- 첫번째, 세번째, 마지막 요소를 추출하시오.  
- True는 리스트의 몇 번째 인덱스인지 추출하시오.

```python
L = [1,0,2,3,0,"A","B",False,[4,5,6]]
print(L[0],L[2],L[-1])

print(L.index(False))
print(L.index(False,2))
print(L.index(False,5))
```

*리스트/튜플에는 find함수를 사용 할 수 없다.  
*index(찾을 요소, 인덱스 번호(이 번호 이후에 가장 처음 나오는 요소가 나오게 된다), 인덱스 번호(이 번호 이전에 나오는 가장 처음 요소)) = 위 범위 내의 가장 처음 나오는 찾는 요소의 인덱스 번호

#연습문제#

#리스트,튜플의 연산자#
리스트와 튜플은 문자여로가 마찬가지로 + 연산을 사용할 수 있습니다.


```python
#튜플의 덧셈
print((1,2,3) + (3,4,5,6))

#리스트의 덧셈
print([1,2,3] + [3,4,5,6])
```

#변경 가능(Mutability)#  
###리스트와 튜플의 차이###  
**리스트는 변경이 가능한 자료형**이지만,  
**튜플은 변경이 불가능한 자료형**이다.  
//문자열도 변경이 불가능한 자료형이다.


```python
L = [0,1,2,3]
T = (0,1,2,3,4)
print(L,T)
L[0] = "X"
print(L) # L의 기존 인덱스0번 요소인 0이 "X"로 변경됨
T[0] ="Y" # T는 튜플로서 요소 변경이 불가능 함
print(T)
```

```python
T = (1,2,3)
print(T)
T[0] = "X"
```

#변경 가능한 리스트의 함수/메소드#
##append## 

**append 메소드**는 **리스트의 맨 마지막에 요소를 추가**한다.  
*형식 : 리스트명 . append ( "추가할 요소")*

```python
X = [1,2,3]
print(X)
X.append(4)
print(X)
```

#연습문제#
기존 학급이 "영희","철수","만수"인 3명으로 구성된 학급이다.  
"영자"가 전학을 왔을 때, "영자"가 포함된 학급 리스트를 추출하시오.

```python
C = ["영희","철수","만수"]
C.append("영자")

print(C)
```

##insert##
**insert는 특정 위치에 값을 추가**한다.  
*형식 : 리스트명 . insert (인덱스번호**(해당인덱스 앞에 놓임)** , 넣을 요소)*

```python
L = [1,3,4,5]
L.insert(1,2)
print(L)
```

#연습문제#
기존 학급이 "영희","철수","만수"인 3명으로 구성된 학급이다.
"영자"가 전학을 왔을 때, "영자"가 "영희"와 "철수" 사이에 포함된 학급 리스트를 추출하시오.

```python
L = ["영희","철수","만수"]
print(L)
L.insert(1,"영자")
print(L)
```

##remove##
리스트에 있는 **특정한 값을 제거**할 때는 **remove**메소드를 사용한다.

```python
L = ["X","Y","Z"]
print(L)
L.remove("Y")
print(L)
```

---

```python
L = [0,1,2,1,3,4]
print(L)
L.remove(1)
print(L)
```

##pop##
**인덱스를 바탕으로 요소를 제거**할 때에는 **pop**을 사용한다.

```python
L = [0,1,2,3,3,3,4,5]
print(L)
L.pop(4)
print(L)
```

---
리스트 . pop () 인덱스를 입력하지 않으면 **맨 뒤**요소가 제거됩니다.

```python
L = [0,1,2,3,4]
print(L)
L.pop()
print(L)
```

##len##
**튜플과 리스트의 길이**는 **len *함수***로 구할수 있다.  
이 때의 **길이**는 **요소의 갯수**이다.

```python
L = ["a","b",1,2]
T = (1,2,3)
print(len(L), len(T))
```

```python
#중첩 리스트의 경우
L = [[1,2,3],[4,5,6]]
print(len(L))
## 퀴즈 : 이 때의 L의 1번째 요소의 길이를 구하시오
print(len(L[0]))
### 퀴즈
L =[[1,2,3,[10,20,30,40,50]],[4,5,6]] #의 10,20,30,40,50 리스트의 길이를 구하시오

print(len(L[0][3]))
```

#변수 할당#
**튜플**은 **여러개의 변수를 동시에 정의할 때 주로 사용**한다.


```python
# 변수1=대입값1, 변수2 = 대입값2, 변수3 = 대입값3
# (변수1, 변수2, 변수3) = (대입값1 ,대입값2, 대입값3)

(a, b, c) = (1,2,3)
print(a)
print(a,b)
print(c)
```

##스와프##
변수 할당을 통해 **변수의 대입값들을 바꿀 수**가 있다 이를 **스와프(swap)**라고 한다.

```python
(a, b) = (1,2)
print(a,b)
(a,b) = (b,a)
print(a,b)
```

---
**튜플은 소괄호를 생략**할 수 있다. 콤마(,)만으로도 표현 가능하다.

```python
T = 1,2,3
print(type(T))
```

---
이를 통해 스와프를 간략히 작성 할 수 있다.

```python
a,b = 1,2
a,b = b,a
print(a, b)
```

##리스트 요소 삭제(del)##

##리스트 확장(extend)##
