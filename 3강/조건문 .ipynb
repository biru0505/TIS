#조건문 if 와 else#
조건이 참(True)일 때, if 이하의 코드 블록을 실행 하고,   
조건이 거짓(False)이면, else 이하의 코드 블록을 실행합니다.

형식 :   
if (불인형 조건블록):  
 (실행블록)  
 else :  
   (실행블록)  

```python
#예시문
T = True
F = False

if T:
  print("참")
else:
  print("참이아님")

if F:
  print("거짓")
else:
  print("거짓이아님 ")
```

#비교 및 논리 연산자#
참/거짓(부울자료형)은 논리연산자로 주로 표현된다.  

연산자| 표현
:-- | --
x < y | x가 y보다작다
x <= y | x가 y보다 작거나 같다
x>y | x가 y보다 크다
x>=y | x가 y보다 크거나 같다
x==y | x와 y가 같다
x!=y | x와 y가 같지 않다.
x in y | x가 y에 포함된다
x not in y | x가 y에 포함되지 않는다
x & y | x와 y가 모두 True
x 합 y |x와 y중 하나가 True
not(x) | x의 부정

```python
print(3 < 2)  #3이 2보다 작은가?
print(3 ==3)  #3과 3은 같은가?
print(2 +2 !=5) # 2+2는 5가 아닌가?
print(len("ABC") >= 2) # "ABC"의 문자열 길이는 2보다 크거나 같은가?
print( 1 in [1,2,3])
print( 1 not in (1,2,3))
print ( (3>1) & (5>4))
print( 3<1 | 5>4)
```

```python
L =[1,2,3]
S = "abcde"
D = { "A" : 1, "B" : 2, "C" : 3}

print(1 in L)
print(2 not in L)
print("abc" in S)
print("ac" in S)
print("A" in D.keys())
print(3 in D.values())
```

##elif문##
**여러개의 조건을 순차적으로 봐야 하는 경우** ***elif(else if의 축약)문***을 사용합니다.

if 조건 1:  
(조건 1)이 참이면 (코드블록 1)출력  
(조건 1)이 거짓이면   
(조건 2)로  

elif(조건 2)  
(조건 2)가 참이면 (코드블록 2)출력  
(조건 2)가 거짓이면  
(조건 3)로  
  
  
elif(조건 3)  
(조건 3)가 참이면 (코드블록 3)출력  
(조건 3)가 거짓이면  
(마지막 조건)로  
  
else (코드블록 4)출력  

```python
L = [2,3,4,5,6]
if (1 in L) and (6 in L):
  print("1과 6이 모두 L안에 있습니다.")
elif (1 in L):
  print("6은 없지만 1은 L안에 있습니다.")
elif (6 in L):
  print("1은 없지만 6은 L안에 있습니다.")
else:
  print("1과 6 모두 L안에 없습니다.")
```

### if/elif/else 사용시 주의점###
if 와 elif/else의 조건을 만족하는 요소는 중복 되어서는 안됩니다. 즉, 서로 조건이 영향을 주는 종속적 관계이기 때문에 if/elif를 사용합니다.  
단, 서로에게 영향을 주지않는 조건들(독립적 관계)은 if/if....를 사용합니다.

```python
D = {"B" : 30}

if "A" not in D.keys():
  D["A"] = 0
if "B" not in D.keys():
  D["B"] = 0
if "C" not in D.keys():
  D["C"] = 0

print(D)
```
