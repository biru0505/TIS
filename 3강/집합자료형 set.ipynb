집합 자료형 : set
**집합 자료형**은 ***순서가 없고* 요소의 *중복을 허용하지 않는***자료형 이다.  
표현은 **중괄호**를 사용하거나, **set 함수**를 사용해 만든다.

```python
S1 = {1,2,3,2,1}
S2 = set([1,2,3,2,1])

print(S1)
print(S2)
```

---
단, 빈집을 만들 때는 {}를 사용할 수 없고,  
set()을 사용해야 한다.
집합 자료형은 중복을 하지 않기에, 리스트와 튜플에 있는 중복 인자를 제거할 때 주로 사용합니다.

```python
L = [1,2,3,1,2,3,]
print(L)
print(set(L)) #해석 : L이라는 리스트를 set함수를 통해 집합화 하여 중복 값을 제거, 이를 추출한다.
```

##집합 연산##
교집합은 **&기호**를 사용하거나 **intersection 메소드**를 사용해서 구할 수 있다.

```python
S1 = {1,2,3,4}
S2 = {3,4,5,6}

print(S1 & S2)
print(S1.intersection(S2))
print(S2.intersection(S1))
```

합집합은 **| 기호**를 사용하거나 **union 메소드**를 사용하여 구할 수 있다.

```python
S1 = {1,2,3,4}
S2 = {3,4,5,6}

print(S1 | S2)
print(S1.union(S2))
print(S2.union(S1))
```

차집합은 **- 기호**를 사용하거나 **difference 메소드**를 사용하여 구할 수 있다.  
*교환법칙 성립안됌*

```python
S1 = {1,2,3,4}
S2 = {3,4,5,6}

print(S1 - S2)
print(S1.difference(S2))
print(S2.difference(S1))
```
