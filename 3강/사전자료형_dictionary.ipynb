#사전 자료형 :Dictionary#
**사전자료형인 딕셔너리(dictionary)**는 **키(key)**와 **값(value)**의 **쌍으로 구성된 순서가 없는 자료형**이다.  
딕셔너리는 **키를 기준으로 데이터를 관리**할 수 있어 자주 사용된다.


키와 값은 콜론(:)으로 대응 시키며, 쌍을 중괄호 안에 넣은 방식으로 딕셔너리를 정의한다.

딕셔너리 값은 사전 자료[키]를 통해 불러올 수 있다.

```python
X = {"A" : 1, "B" : 2, "C" : 3}
## 1.대응 시키는 방법                       {키(key) : 값(value) } 
## 2.키(key)로 값(value)를 불러 오는 방법   X["A"] = 1

print(X["B"])
```

##새로운 쌍 추가하기 ##
새로운 쌍의 키(기존 키 값과 중복되면 안됌) = value 형식으로 추가한다


##기존 쌍 변경하기##
바꾸고 싶은 키값을 호출하여 새로운 value 값을 추가한다

##기존 쌍 제거하기##
del(쌍["키"])를 통해 제거할 수 있다.


```python
X = {"A" : 1, "B" : 2, "C" : 3}
print(X)
X["D"] = 4  #쌍 추가
print(X)
X["A"] = 7  #쌍 변경
print(X)
del(X["B"])  #쌍 제거
print(X)
type(X)
```

#연습문제
여러분의 이름(name)과 거주하는 동이름(dong)과 좋아하는 색깔(color)를 딕셔너리 변수로 만들어 보세요

```python
my_profile = {"name" :"최관수", "dong" : "논현동", "color" : "black"}

print(my_profile)

# 기존 프로필에 좋아하는 숫자를 추가하시오.

my_profile ["number"] = "5"

print(my_profile)

# 잘못 기입한 이름을 옳게 수정하시오.
my_profile["name"] = "채관수"
print(my_profile)

# 이름을 삭제하시오
del(my_profile["name"])
print(my_profile)

# 좋아하는 운동 2가지를 "favorite_sports"라는 키로 추가해 보세요.
my_profile["favorite_sports"] = "농구", "야구" #{1 : "농구", 2 : "야구"} #딕셔너리 형태, 리스트형,튜플형 모두 가능
print(my_profile)
my_profile["favorite_sports"][1]
type(my_profile["favorite_sports"])
```

**딕셔너리의 구성 요소를 확인하는 메소드**로는  
keys()  모든 키  
values()모든 값  
items() 모든 쌍  
이 있습니다. 

```python
X = {"A" : 1, "B" : 2, "C" : 3}
print(X.keys())
print(X.values())
print(X.items())
```
